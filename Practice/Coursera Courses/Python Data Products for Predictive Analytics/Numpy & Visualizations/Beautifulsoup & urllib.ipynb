```python
from urllib.request import urlopen
```

```python
# Example I wanna collect data from this URL
# https://www.goodreads.com/book/show/5671.The_Ethics_of_Rhetoric
```

### Step 1: Extract the html code of the webpage

```python
file = urlopen('https://www.goodreads.com/book/show/5671.The_Ethics_of_Rhetoric')
html = str(file.read())
html
```

### Step 2: Find the relevant parts of the code that correspond to what you are trying to extract

```python
# Split at the place where you find <div id ="review_
# Ignore the first split because it consists of stuff that are before this portion you identified
reviews = html.split('<div id="review_')[1:]
len(reviews)
```

```python
def parseReview(review):
    reviews = {}
    reviews['stars'] = review.split('<span class="staticStars" title="')[1].split('"')[0]
    reviews['date'] = review.split('<a class="reviewDate')[1].split('<')[0]
    reviews['user'] = review.split('<a title="')[1].split('"')[0]
    
    shelves = []
    try:
        shelfBlock = review.split('<div class="uitext greyText bookshelves">')[1].split('</div')[0]
        for s in shelfBlock.split('shelf=')[1:]:
            shelves.append(s.split('"')[0])
        reviews['shelves'] = shelves
    
    except Exception as e:
        pass
    
    reviewBlock = review.split('<div class="reviewText stacked">')[1].split('</div')[0]
    reviews['reviewBlock'] = reviewBlock
    
    return reviews
```
