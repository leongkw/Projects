```python
# Testing of menu buttons

# Basically a 'handler' function like the echo bot example above
# When message is received, it goes through a series of if-else statements to determine what to do
# \n is for the displayed message to go next line
welcome_text = "Welcome to my fruit shop.\n\nI am FruitBot and I will be your service assistant.\n\nPlease type /fruit to choose your fruit."

def on_chat_message(message):
    # Retrieves content_type, chat_type, sender_id from message object
    content_type, chat_type, sender_id = telepot.glance(message)
    
    # Type /start and bot will send the welcome text
    if message['text'] == '/start':
        bot.sendMessage(sender_id, welcome_text)
    
    # Type /fruit and it will display the button menu showing all the fruits
    if message['text'] == '/fruit':
        # Creates a menu with buttons in the chat
        # For every button you want, you need to create a [InlineKeyboardButton()]
        # text='' will be the displayed text for the button
        # callback_data i presume is what will be returned to you when user presses it
        keyboard = InlineKeyboardMarkup(inline_keyboard=[[InlineKeyboardButton(text='Apple', 
                                                          callback_data='Apple')],
                                                         [InlineKeyboardButton(text='Orange',
                                                          callback_data='Orange')],
                                                         [InlineKeyboardButton(text='Watermelon',
                                                          callback_data='Watermelon')]
                                        ])
    
        # To show the keyboard, in sendMessage(), you need to specify reply_markup=keyboard
        bot.sendMessage(sender_id, 'Please choose your fruit:', reply_markup=keyboard)
        print(message)

# 'Handler' function to handle button presses from the menu buttom
def on_callback_query(message):
    query_id, from_id, query_data = telepot.glance(message, flavor='callback_query')
    recipient = message['message']['chat']['id']
    print('Callback Query:', query_id, from_id, query_data)
    
    # .answerCallbackQuery() provides a text response when you click on the button
    # Response type is a tooltip style message that will fade after a certain amount of time
    # Note that this doesn't produce a text response back
    if query_data == 'Apple':
        bot.answerCallbackQuery(query_id, text='Ahhh wise choice. An apple a day keeps the doctor away.')
        # You can send a text message back on button presses with bot.sendMessage
        bot.sendPhoto(recipient, photo='https://imgur.com/a/Hm777Ea', caption='Bright red juicy apple') 
        
    if query_data =='Orange':
        bot.answerCallbackQuery(query_id, text='Oranges are not only sweet but they are also rich in Vitamin C!')
        bot.sendPhoto(recipient, photo='https://imgur.com/a/9Rnsepm', caption='Sweet succulent orange. Perfect for the summer.')
    
    if query_data =='Watermelon':
        bot.answerCallbackQuery(query_id, text='Refreshing watermelon, full of goodness.')
        bot.sendPhoto(recipient, photo='https://imgur.com/a/AAeJNY6', caption="Juicy watermelon to keep your body fresh and hydrated.")

# Bot Token
Token = 'insert bot token here'

# Creating the bot
bot = telepot.Bot(Token)

# MessageLoop to keep running getUpdates, except in this case
# A dictionary is passed in format of {'source' : handler_function, ...}
MessageLoop(bot, {'chat' : on_chat_message,
                  'callback_query' : on_callback_query}
           ).run_as_thread()
```
