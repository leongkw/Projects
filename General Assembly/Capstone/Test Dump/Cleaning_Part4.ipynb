```python
import pandas as pd
```

```python
stats = pd.read_csv('stats_1.4.csv')
stats.drop(labels='Unnamed: 0', axis=1, inplace=True)
```

```python
stats.head(50)
```

```python
mvp = pd.read_csv('mvp.csv')
```

```python
mvp = mvp[mvp['Season'] > 1977]
mvp.reset_index(inplace=True)
```

```python
mvp.drop(labels='index', axis=1, inplace=True)
```

```python
MVP = pd.Series([])

for i in range(len(stats)):
    for j in range(len(mvp)):
        if stats['Year'][i] == mvp['Season'][j] and stats['Player'][i] == mvp['Player'][j]:
            MVP[i] = 1
            
```

```python
MVP.value_counts()
```

```python
stats.insert(1, 'MVP', MVP)
```

```python
stats['MVP'].fillna(0, inplace=True)
```
